```python
df = pd.DataFrame.from_dict(categorias)
idCategoria= df.iloc[4]['id']
df.shape
```

```python
urlCat = "https://api.mercadolibre.com/categories/"+idCategoria
urlCat
respuesta=requests.get(urlCat)
respuesta.json()
```

```python
columnas = ['idCategoria', 'Categoria', 'idSubCategoria', 'subCategoria', 'noItems']
dfd=pd.DataFrame(columns=columnas)
dfd
```

```python
for c in categorias:
  print(c['id'], c['name'])
  urlCat = "https://api.mercadolibre.com/categories/" + idCategoria 
  #urlCat
  respuesta=requests.get(urlCat)
  contenido= respuesta.json()
  subcat=contenido ['children_categories']
  for sc in subcat:
    datos={'idCategoria':c['id'], 'Categoria':c['name'],
           'idSubCategoria': sc['id'], 'SubCategoria': sc['name'],
           'noItems': sc['total_items_in_this_category']}
    dfd=dfd.append(datos, ignore_index=True) 
```
