```python
columnas = ['idCategoria', 'Categoria', 'idSubCategoria', 'subCategoria', 'noItems']
dfd=pd.DataFrame(columns=columnas)
dfd
```

```python
for c in categorias:
  print(c['id'], c['name'])
  urlCat = "https://api.mercadolibre.com/categories/" + idCategoria 
  #urlCat
  respuesta=requests.get(urlCat)
  contenido= respuesta.json()
  subcat=contenido ['children_categories']
  for sc in subcat:
    datos={'idCategoria':c['id'], 'Categoria':c['name'],
           'idSubCategoria': sc['id'], 'SubCategoria': sc['name'],
           'noItems': sc['total_items_in_this_category']}
    dfd=dfd.append(datos, ignore_index=True)
```

```python
url_countries='https://restcountries.com/v3.1/region/europe'
form_www=''
headers={}
respuesta_countries=requests.request("GET",url_countries,headers=headers,data=form_www)
respuesta_countries.text
```

```python
json_respuesta_countries=respuesta_countries.json()
print(json_respuesta_countries)
```

```python
columnas = ['NombreComun', 'Subregion', 'Poblacion']
dfc=pd.DataFrame(columns=columnas)
dfc
```

```python
for c in json_respuesta_countries:
    for m in c['currencies']:
        datos_countries={'NombreComun':c['name']['common'], 'Subregion':c['subregion'], 'Poblacion': c['population'], 'Moneda':c['currencies'][m]['name']}
        dfc=dfc.append(datos_countries, ignore_index=True)
```
