```python
#se importan las librerías a utilizar
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import csv
```

```python
# Define URL a descargar y hace la petición HTTP
url = 'http://www.diputados.gob.mx/LeyesBiblio/index.htm'
response = requests.get(url).text
soup = BeautifulSoup(response)
```

```python
#Extraemos la tabla que se puede visualizar en la página
tabla = soup.find('table')
```

```python
tabla_arreglo=tabla.find_all("tr")
nrenglones = len(tabla_arreglo)-1
print("Numero de renglones: ",nrenglones)
datos = {'Numero':'', 'Ley':'', 'Ultima_reforma':'', 'Texto_vigente':''}
quitar = "\r\n\t"
nombre_archivo ="Leyes_Federales.csv"
with open(nombre_archivo, 'a') as archivo_csv:
  writer = csv.writer(archivo_csv)
  writer.writerow(['Numero', 'Ley', 'Ultima_reforma', 'Texto_vigente'])
  for r in range(1, nrenglones+1):
    celda=tabla_arreglo[r].find_all('td')
    for c in range(0,3):
      textocelda=celda[c].text.rstrip()
      for x in range(len(quitar)):
        textocelda = textocelda.replace(quitar[x],"")
      if c == 0:
        datos['Numero'] = textocelda
      if c == 1:
        datos['Ley'] = textocelda
      if c == 2:
        datos['Ultima_reforma'] = textocelda
    for href in tabla_arreglo[r].find_all('a',href=True):
      celda_link=href.get('href')
      datos['Texto_vigente']=celda_link
    print(datos)
    writer.writerow([datos['Numero'],datos['Ley'],datos['Ultima_reforma'], datos['Texto_vigente']])
```
