```python
import urllib
from bs4 import BeautifulSoup
```

```python
#url puede ser cualquier pagina html
url='https://github.com/jorgejoshuatt/Adquisicion-de-datos'
html=urllib.request.urlopen(url)
html
```

```python
soup=BeautifulSoup(html)
soup
```

```python
tags=soup('a')
tags
```

```python
print('Enlaces de la pagina principal: \r\n')
for tag in tags:
  try:
    print(tag) #tag.content[0], tag.get('href')
  except:
    print('algo ocurrio')
```

```python
print('\r\n Enlaces de las páginas secundarias:\r\n')
for tag in tags:
  nevur1=tag.get('href')
  print('Accediendo a los enlaces dentro de la pagina', nevur1)
  try:
    if nevur1[0:4]=='http':
      html2=urllib.request.urlopen(nevur1)
      soup=BeautifulSoup(html2)
      tags=soup('a')
      print('enlaces en la pagina principal: \r\n')
      for tag in tags:
        print(tag) #tag.get('href')
    else:
      html2=urllib.request.urlopen(url+nevur1)
  except:
    print('Algo ocurrio')
```
