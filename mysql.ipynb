```python
import datetime
sql = "SELECT clave_alu, CONCAT_WS(' ', ap_paterno, ap_materno, nombre) alumno, sexo, curp, fedita " + "FROM alumnos " + "WHERE fedita BETWEEN %s AND %s " + "ORDER BY 5"

fini = datetime.date(2016, 11, 1)
ffin = datetime.date(2016, 11, 30)
cursor.execute(sql, (fini, ffin))
rs = cursor.fetchall()
print(len(rs))
for r in rs: 
  print(r)
```

```python
autor = 'JORGE JOSHUA TREJO MARTINEZ' 
sql = "SELECT '" + autor + "' as autor, a.clave_alu, \
CONCAT_WS(' ', ap_paterno, ap_materno, nombre) alumno, \
IFNULL(SUM(p.pago), 0) AS tpago, COUNT(p.pago) AS npago \
FROM alumnos a LEFT JOIN pagos p ON(a.clave_alu = p.clave_alu) \
GROUP BY a.clave_alu \
ORDER BY 2"
cursor.execute(sql)
rs = cursor.fetchall()
print(len(rs))
for r in rs:
  print(r)
```

```python
df = pd.read_sql_query(sql = sql, con = conn)
df.info()
```
